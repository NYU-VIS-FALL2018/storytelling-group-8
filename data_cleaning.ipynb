```python
import pandas as pd
from itertools import groupby 
from collections import OrderedDict
import json 
```

```python
df = pd.read_csv('GlobalLandTemperaturesByCity.csv', dtype={
        "Year" : str,
        "Month" : str,
        "AverageTemperature" : float,
        "City" : str,
        "Country" : str,
        "Latitude" : str,
        "Longitude" : str,
    })

results = []
```

```python
def convert_latlong(v):
    value=float(v[:-1])
    direction=v[-1]
    if direction=='S' or direction == 'W':
        return str(-1*(value))
    return str(value)
```

```python
df= df.dropna()
```

```python
df.Latitude=df.Latitude.apply(convert_latlong)
df.Longitude=df.Longitude.apply(convert_latlong)
```

```python
df.tail(100)
```

```python

finalList = []
finalDict = {}
grouped = df.groupby(['City', 'Country','Latitude','Longitude'])
for key, value in grouped:

    dictionary = {}
    month_dict={}
    j = grouped.get_group(key).reset_index(drop=True)
    dictionary['City'] = j.at[0, 'City']
    dictionary['Country'] = j.at[0, 'Country']
    dictionary['Latitude'] = float(j.at[0, 'Latitude'])
    dictionary['Longitude'] = float(j.at[0, 'Longitude'])

    d={}
    dictList = {}
    anotherDict = {}
    
    for i in j.index:
        dictList.update( {j.at[i, 'Year'] : j.at[i, 'AverageTemperature']} ) 
        
        month_dict.update({j.at[i, 'Month']: dictList})
    dictionary['Month'] = month_dict
    
    anotherDict.update({j.at[0, 'Month']: dictList})
    
    finalList.append(dictionary)
# print(finalList)
with open('data.json', 'w') as outfile:
    json.dump(finalList, outfile)
```
